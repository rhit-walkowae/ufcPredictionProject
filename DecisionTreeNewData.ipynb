```python
Their_category = dfTheir.select_dtypes(exclude=np.number)
Their_numerical = dfTheir.select_dtypes(include=np.number)
Opp_category = dfOpp.select_dtypes(exclude=np.number)
Opp_numerical = dfOpp.select_dtypes(include=np.number)
```

```python
dfnew['Fighters_Stance'] = Their_category
dfnew['Opponent_Stance'] = Opp_category
dfnew.head(3)
```

```python

for cols in Their_numerical.columns:
    #print(cols)
    label = cols.removeprefix('THEIR')
    label = 'DIFF' + label
    index = Their_numerical.columns.get_loc(cols)
    colOpp = Opp_numerical.columns[index]
    dfnew[label] = Their_numerical[cols] - Opp_numerical[colOpp]
```
